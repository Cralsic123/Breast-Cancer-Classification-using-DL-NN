```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
X = data_frame.drop(columns='label', axis=1) # adding all columns except labels by dropping column "label"
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Now we are going to standardize the data by scaling it and transforming it.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

Here I am going to make the neural network. Well honestly I am doing this to brish my basics since the second year, So I guess we shall pull off with neural networks. One input layer for features and one output layer and then multiple hidden layers inside.

We will use two libraries.

Tensorflow and Keras

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3) # to provide same accuracy everytime we run the program
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
]) # input shape is 30
# 3rd one is the output layer
# we can se previously that the X_train is consisting of 569 X 30 columns
# we need to flatten it, such that into a single dimensional one
# 3rd is the output, we have two classes so only 2
```

```python
# compiling the Neural Network
# if we were using one hot encoding then we would have used categorical_crossentropy
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Now we will be looking into the plots

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Now I am trying on a test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
